```python
import skimage
import matplotlib.pyplot as plt
```

```python
images = skimage.io.imread_collection("test/*.jpeg")
```

```python
len(images)
```

```python
for j,image in enumerate(images):
    new = skimage.transform.resize(image,(299,299))
    skimage.io.imsave(f"test/test299/image_{j}.jpeg",new)
```

```python
imag = skimage.io.imread("data/RGB/test/image_48.jpeg")
imag1 = skimage.transform.rotate(imag,45)
plt.imshow(imag1)
skimage.io.imsave("data/RGB/test/new3.jpeg",imag1)
```
